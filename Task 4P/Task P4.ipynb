#### First and foremost , before any analysis, I have to import relevant packages to easily implement and visualise the analysis more effectively.Originally, some columns are not being in the correct calculated systems format such as SI or derived units so I just converted all of the entries of each column into the right format for the purpose of the analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("weather.csv")

print("Original Dataframe:")
display(df)
# Convert the temp and dewp column from Fahrenheit into Celsisus degree
df[["temp","dewp"]] = (df[["temp","dewp"]] - 32 ) * 5/9

# Convert the precip column from inches into milimeters
df["precip"] = df["precip"] * 25.4

# Convert the vsib column from miles into meters
df["visib"] = df["visib"] * 1609.34 

# Convert the wind_speed , wind_gust column from mph to m/s
df[["wind_speed","wind_gust"]] = df[["wind_speed","wind_gust"]] * 0.44704

#Display the new DataFrame
print("Converted Dataframe:")
display(df)

```

#### From this code snippet , I tried to convert all the columns format to be more appropriate.Specifically, the temperature columns ("temp" and "dewp") are converted from Fahrenheit to Celsius using the formula (F -32) * 5/9 = C , the "precip" column, representing precipitation in inches, is converted to millimeters by multiplying by 25.4 (since 1 inch is approximately equal to 25.4 millimeters).
#### The "visib" column, representing visibility in miles, is converted to meters by multiplying by 1609.34 (since 1 mile is approximately equal to 1609.34 meters), The wind speed columns ("wind_speed" and "wind_gust") are converted from miles per hour to meters per second by multiplying by 0.44704 (since 1 mile per hour is approximately equal to 0.44704 meters per second).

 

#### The next phase is finding the daily mean wind speeds in the LGA airport for each day.Basically in each specific day in one year it also has multiple values in terms of different hours. From that we can calculate the daily mean values of the wind speeds.

```python
# Filter out the airport 
df_LGA = df[df["origin"] == "LGA"]
print("Dataframe which only contains LGA rows:")
display(df_LGA)

#Then I group by the the year,month,day to calculate the average wind speed in each day
df_LGA = df_LGA.groupby(["year","month","day"])[["wind_speed"]].mean().reset_index()
print("The Data Frame which represents the average of the wind speed on each day:")
display(df_LGA)

```

#### Regarding this code fragment , at first I filtered out the entries which only represent the "LGA" airport.Then I proceeded to calculate the daily mean wind speeds for the LGA airport. It achieved this by grouping the DataFrame 'df' by the 'Date' column using the .groupby() method in terms of the Year,Month and Day and then applying the .mean() function to compute the mean wind speed for each day.
#### Finally, the resulting daily mean wind speeds for the LGA airport are displayed, providing insights into the average wind conditions observed over each day throughout the dataset. This process enables the analysis of wind patterns and variations, facilitating better understanding and interpretation of the data.

#### To illustrate the daily fluctuations of wind speeds based on the latest calculations, I've enhanced the plot's aesthetic appeal. This involved crafting a new 'Date' column while removing unnecessary data columns to ensure clarity. Subsequently, I refined and adjusted the x-axis and y-axis labels to enhance readability and comprehension.

```python
# Format the df_LGA for plotting in an aesthetic way

#Create a new column named Dtae combined the Year ,Month and Day
df_LGA["Date"] = pd.to_datetime(df_LGA[["year","month","day"]])
#Dropping all the irrelevant columns
df_LGA.drop(columns = ["year","month","day"],inplace = True)
#Set the index of the dataframe as the Date (it will be denoted as the x axis)
df_LGA.set_index("Date",inplace = True)

print("Modified LGA Dataframe with Date:")
display(df_LGA)
# Plot a graph
df_LGA.plot()

# Modify the title and x,y labels
plt.legend().remove()
plt.title("Daily Average Wind Speeds in 2023")
plt.xlabel("Date")
plt.ylabel("Daily Average wind speed (m/s) at LGA")
plt.xticks(rotation=45)
```

#### The above code begins by creating a new "Date" column by combining the "Year", "Month", and "Day" columns and converting them into a datetime format. Irrelevant columns are then dropped to declutter the DataFrame.
#### The "Date" column is set as the index, which will represent the x-axis when plotting. The data is plotted, and to enhance visual appeal, the legend is removed. The title is modified to "Daily Average Wind Speeds in 2023", and the x and y-axis labels are adjusted accordingly. Additionally, the x-axis labels are rotated by 45 degrees for better readability.

#### The final step of this report involves identifying the top 10 windiest dates based on their corresponding wind speed values. Subsequently, the wind speed values for each entry are rounded to a precision of 2 decimal points. Finally, a DataFrame representing the top 10 windiest dates, along with their rounded wind speed values, is displayed.

```python
# Sorting the values in the descending order which means finding the top windiest days
df_LGA_top10 = df_LGA.sort_values(by = "wind_speed",ascending = False)

#Rounding all the entries for the data looks better
df_LGA_top10["wind_speed"] = df_LGA_top10["wind_speed"].round(2)
print("LGA Dataframe with daily mean wind speeds from highest to lowest")
display(df_LGA_top10)
#Extract top 10 windiest Date together with its wind speed figures 
print("Top 10 windiest Date as well as its corresponded wind speed:")
display(df_LGA_top10.iloc[:10])
```

#### From this code snippet , it sorts the DataFrame df_LGA by the "wind_speed" column in descending order to find the top windiest days. The sorted DataFrame is stored in df_LGA_top10.Then  it rounds all entries in the "wind_speed" column of the DataFrame df_LGA_top10 to two decimal places for improved readability. Last but not least , extracting the top 10 windiest dates along with their corresponding wind speed figures from the sorted DataFrame df_LGA_top10 using the iloc indexer, and displaying the result.
